# Simulated Data Experiment
In this notebook, the performance of recurrent neural networks is evaluated on the simulated time-series data.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data
There are three different simulated data, each of which represents a 1D time-series:
1. **Henon**: from *---*; a time-series of length 16384 in range [-1.798490; 1.782120] with mean of 0.365823 and standard deviation of 0.005009.
2. **Rossler**: from *---*; a time-series of length 8192 in range [-14.592151; 17.294220] with mean of 0.128972 and standard deviation of 7.971135.
3. **Lorenz**: from *---*; a time-series of length 16384 in range [-18.793848; 18.097797] with mean of 0.312581 and standard deviation of 7.918619.

```python
data_henon = pd.read_csv("../data/simulated/henon.csv", header=None)
data_henon.describe()
```

```python
data_rossler = pd.read_csv("../data/simulated/rossler.csv", header=None)
data_rossler.describe()
```

```python
data_lorenz = pd.read_csv("../data/simulated/lorenz.csv", header=None)
data_lorenz.describe()
```

```python
idx_henon = data_henon.sample(3).index.values
idx_rossler = data_rossler.sample(3).index.values
idx_lorenz = data_lorenz.sample(3).index.values
```

```python
f, axarr = plt.subplots(3, 3, figsize=(15,9))
f.suptitle("Examples of the simulated data")
for i in range(len(idx_henon)):
    axarr[0, i].plot(data_henon[idx_henon[i]:idx_henon[i]+100]);
    axarr[0, i].set_title("Henon: Example " + str(i+1));
    axarr[1, i].plot(data_rossler[idx_rossler[i]:idx_rossler[i]+500]);
    axarr[1, i].set_title("Rossler: Example " + str(i+1));
    axarr[2, i].plot(data_lorenz[idx_lorenz[i]:idx_lorenz[i]+500]);
    axarr[2, i].set_title("Lorenz: Example " + str(i+1))
f.subplots_adjust(hspace=0.3)
```
